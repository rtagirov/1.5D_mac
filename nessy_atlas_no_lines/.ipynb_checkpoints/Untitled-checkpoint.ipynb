```python
import os

def num_lines(f): return sum(1 for line in open(f))

def read_popnum(path):

    srcdir = os.getcwd()

    os.chdir(path)

    os.system('grep -w LEVEL datom.inp > levels.out')

    dpn = num_lines('atm.inp')

    ln = num_lines('levels.out')

    rne = np.zeros(dpn)

    popnum = np.zeros((ln, dpn))

    dpn3 = np.ceil(dpn / 3.0)

    dpnln3 = int(dpn * ln / 3)

    head = int(10 + 2 * dpn3)

    os.system('head -n ' + str(head) + ' POPNUM | tail -n ' + str(int(dpn3)) + ' > rne.out')

    head = int(head + 1 + dpnln3)

    os.system('head -n ' + str(head) + ' POPNUM | tail -n ' + str(int(dpnln3)) + ' > popnum.out')

    lev = []

    f = open('levels.out', 'r')

    for line in f:

        parts = line.split()

        lev_name = parts[1]

        if (lev_name[0] == 'H'): lev_name = parts[1] + parts[2]
            
        lev_name = lev_name.replace('-', '')

        if (lev_name[0 : 2] == 'HI' or lev_name[0 : 2] == 'HM' or lev_name[0 : 4] == 'HEII'):

            lev_name = lev_name.replace('.', '')

        if (lev_name == 'HMINUS1'): lev_name = 'HMINUS'

        lev.append(lev_name)

    f.close()

    f = open('rne.out', 'r')

    i = 0

    for line in f:

        parts = line.split()

        for part in parts:

            rne[i] = float(part)

            i = i + 1

    f.close()

    f = open('popnum.out', 'r')

    i = 0; j = 0

    for line in f:

        parts = line.split()

        for part in parts:

            popnum[j, i] = float(part)
            
            i = i + 1

            if i != 0 and i % dpn == 0:

                i = 0

                j = j + 1

    f.close()

    os.system('rm levels.out rne.out popnum.out')

    os.chdir(srcdir)

    return lev, rne, popnum
```

```python
Ia = np.zeros((5, 119))
In = np.zeros((5, 120))

Ia_itrp = np.zeros((5, 120))

wa = np.loadtxt('./mpsa.wave.180.300')
wn = np.arange(1805, 3005, 10) / 10.0

angles = [10, 8, 6, 4, 2]

mu = ['1.0', '0.8', '0.6', '0.4', '0.2']

for i in range(len(angles)):

    Ia[i, :] = np.loadtxt('./' + str(angles[i]) + '/mpsa.punch.180.300')
    In[i, :] = np.loadtxt('./' + str(angles[i]) + '/spec.out')
```

```python
import matplotlib.pyplot as plt

from scipy import interpolate

import itertools

fig, ax = plt.subplots(nrows = 5, ncols = 2, figsize = (20, 25))

for i in range(len(angles)):

    f = interpolate.interp1d(wa, Ia[i, :])

    Ia_itrp[i, :] = f(wn)

    ax[i, 0].plot(wa, Ia[i, :], color = 'black', label = 'atlas')
    ax[i, 0].plot(wn, In[i, :], color = 'red',   label = 'nessy')

    ax[i, 1].plot(wn, Ia_itrp[i, :] / In[i, :], color = 'black')
    ax[i, 1].axhline(y = 1.0, linestyle = '--', color = 'black')

    ax[i, 0].set_yscale('log')
    ax[i, 1].set_yscale('log')
    
    ax[i, 0].set_ylim(1e-9, 1e-4)
    ax[i, 1].set_ylim(1e-2, 1e+1)

    ax[i, 0].set_ylabel('$\mu$ = ' + mu[i] + ': Intensity, cgs', fontsize = 20)
    ax[i, 1].set_ylabel('atlas / nessy', fontsize = 20)
    
for p, q in itertools.product(range(5), range(2)): ax[p, q].set_xlim(180, 300)

ax[4, 0].set_xlabel('Wavelength, nm', fontsize = 20)
ax[4, 1].set_xlabel('Wavelength, nm', fontsize = 20)

leg = ax[0, 0].legend(framealpha = 1, loc = 4, handletextpad = 1, prop = {'size': 15.5})

plt.savefig('continuum.pdf', bbox_inches = 'tight')

plt.show()
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (10, 5))

for i in range(len(angles)):

    f = interpolate.interp1d(wa, Ia[i, :])

    Ia_itrp[i, :] = f(wn)

    ax.plot(wn, Ia_itrp[i, :] / In[i, :], label = '$\mu$ = ' + mu[i])
    
ax.axhline(y = 1.0, linestyle = '--', color = 'black')
    
ax.set_ylim(0, 9)

ax.set_ylabel('atlas / nessy', fontsize = 20)
    
ax.set_xlim(180, 300)

ax.set_xlabel('Wavelength, nm', fontsize = 20)

leg = ax.legend(framealpha = 1, loc = 1, handletextpad = 1, prop = {'size': 15.5})

plt.savefig('cont_ratio.pdf', bbox_inches = 'tight')

plt.show()
```

```python
fig, ax = plt.subplots(nrows = 5, ncols = 2, figsize = (20, 25))

k = 1.380649e-16

for i in range(len(angles)):
    
    h = np.loadtxt('./' + str(angles[i]) + '/atm.inp', usecols = [0])
    T = np.loadtxt('./' + str(angles[i]) + '/atm.inp', usecols = [1])
    n = np.loadtxt('./' + str(angles[i]) + '/atm.inp', usecols = [2]) / k / T
    
    x, rne, popnum = read_popnum(str(angles[i]))
    
    ne_nes = n * rne
    
    hm_nes = n * popnum[0, :]
    
    ax[i, 0].plot(h, ne_nes)
    ax[i, 1].plot(h, hm_nes)
    
    ax[i, 0].set_yscale('log')
    ax[i, 1].set_yscale('log')
    
    ax[i, 0].set_xlim(0, max(h))
    ax[i, 1].set_xlim(0, max(h))
    
plt.show()
```
