```python
import os
import numpy as np
import matplotlib.pyplot as plt
import glob
import sys

from tqdm import tqdm

def frac(a, rand_idx, i):
    
    a_frac_n = np.zeros((int(512 * 512 / 2**i), 120))
    a_frac_r = np.zeros((int(512 * 512 / 2**i), 120))
    
    a_frac_n[:, :] = a[0::2**i, :]
    
    a_frac_r[:, :] = a[rand_idx, :]
    
    return a_frac_n, a_frac_r

def mean_and_std(a):
    
    m = np.mean(a)
    
    s = np.std(a)
    
    return m, s

I0nf = np.load('specs_3D.npz')['I0n'].reshape(512 * 512, 120)
ISnf = np.load('specs_3D.npz')['ISn'].reshape(512 * 512, 120)
I3nf = np.load('specs_3D.npz')['I3n'].reshape(512 * 512, 120)

I0lf = np.load('specs_3D.npz')['I0l'].reshape(512 * 512, 120)
ISlf = np.load('specs_3D.npz')['ISl'].reshape(512 * 512, 120)
I3lf = np.load('specs_3D.npz')['I3l'].reshape(512 * 512, 120)

Nfrac = 7

I0n_m_n = np.zeros((Nfrac, 120))
ISn_m_n = np.zeros((Nfrac, 120))
I3n_m_n = np.zeros((Nfrac, 120))
I0l_m_n = np.zeros((Nfrac, 120))
ISl_m_n = np.zeros((Nfrac, 120))
I3l_m_n = np.zeros((Nfrac, 120))

I0n_d_n = np.zeros((Nfrac, 120))
ISn_d_n = np.zeros((Nfrac, 120))
I3n_d_n = np.zeros((Nfrac, 120))
I0l_d_n = np.zeros((Nfrac, 120))
ISl_d_n = np.zeros((Nfrac, 120))
I3l_d_n = np.zeros((Nfrac, 120))

I0n_m_r = np.zeros((Nfrac, 120))
ISn_m_r = np.zeros((Nfrac, 120))
I3n_m_r = np.zeros((Nfrac, 120))
I0l_m_r = np.zeros((Nfrac, 120))
ISl_m_r = np.zeros((Nfrac, 120))
I3l_m_r = np.zeros((Nfrac, 120))

I0n_d_r = np.zeros((Nfrac, 120))
ISn_d_r = np.zeros((Nfrac, 120))
I3n_d_r = np.zeros((Nfrac, 120))
I0l_d_r = np.zeros((Nfrac, 120))
ISl_d_r = np.zeros((Nfrac, 120))
I3l_d_r = np.zeros((Nfrac, 120))

for i in range(Nfrac):
    
    rand_idx = np.random.choice(512 * 512, size = int(512 * 512 / 2**i), replace = False)
    
    I0n_frac_n, I0n_frac_r = frac(I0nf, rand_idx, i)
    ISn_frac_n, ISn_frac_r = frac(ISnf, rand_idx, i)
    I3n_frac_n, I3n_frac_r = frac(I3nf, rand_idx, i)
    
    I0l_frac_n, I0l_frac_r = frac(I0lf, rand_idx, i)
    ISl_frac_n, ISl_frac_r = frac(ISlf, rand_idx, i)
    I3l_frac_n, I3l_frac_r = frac(I3lf, rand_idx, i)
    
    for j in tqdm(range(120), desc = 'frac = 1/' + str(int(2**i))):

        I0n_m_n[i, j], I0n_d_n[i, j] = mean_and_std(I0n_frac_n[:, j])
        ISn_m_n[i, j], ISn_d_n[i, j] = mean_and_std(ISn_frac_n[:, j])
        I3n_m_n[i, j], I3n_d_n[i, j] = mean_and_std(I3n_frac_n[:, j])
        
        I0n_m_r[i, j], I0n_d_r[i, j] = mean_and_std(I0n_frac_r[:, j])
        ISn_m_r[i, j], ISn_d_r[i, j] = mean_and_std(ISn_frac_r[:, j])
        I3n_m_r[i, j], I3n_d_r[i, j] = mean_and_std(I3n_frac_r[:, j])
        
        I0l_m_n[i, j], I0l_d_n[i, j] = mean_and_std(I0l_frac_n[:, j])
        ISl_m_n[i, j], ISl_d_n[i, j] = mean_and_std(ISl_frac_n[:, j])
        I3l_m_n[i, j], I3l_d_n[i, j] = mean_and_std(I3l_frac_n[:, j])
        
        I0l_m_r[i, j], I0l_d_r[i, j] = mean_and_std(I0l_frac_r[:, j])
        ISl_m_r[i, j], ISl_d_r[i, j] = mean_and_std(ISl_frac_r[:, j])
        I3l_m_r[i, j], I3l_d_r[i, j] = mean_and_std(I3l_frac_r[:, j])
```

```python
w = np.arange(1805, 3005, 10) / 10

plt.close('all')

fig, ax = plt.subplots(nrows = 3, ncols = 4, figsize = (15, 10))

fig.suptitle('NLTE', y = 1.005)

fig.tight_layout(w_pad = 3)

for k in [0, 2]: ax[0, k].set_title('nonrandom')
for k in [1, 3]: ax[0, k].set_title('random')

for i in range(1, Nfrac):
    
    ax[0, 0].plot(w, I0n_m_n[i, :] / I0n_m_n[0, :], label = '1/' + str(int(2**i)))
    ax[1, 0].plot(w, ISn_m_n[i, :] / ISn_m_n[0, :])
    ax[2, 0].plot(w, I3n_m_n[i, :] / I3n_m_n[0, :])
    
    ax[0, 1].plot(w, I0n_m_r[i, :] / I0n_m_r[0, :], label = '1/' + str(int(2**i)))
    ax[1, 1].plot(w, ISn_m_r[i, :] / ISn_m_r[0, :])
    ax[2, 1].plot(w, I3n_m_r[i, :] / I3n_m_r[0, :])
    
    ax[0, 2].plot(w, I0n_d_n[i, :] / I0n_d_n[0, :], label = '1/' + str(int(2**i)))
    ax[1, 2].plot(w, ISn_d_n[i, :] / ISn_d_n[0, :])
    ax[2, 2].plot(w, I3n_d_n[i, :] / I3n_d_n[0, :])
    
    ax[0, 3].plot(w, I0n_d_r[i, :] / I0n_d_r[0, :], label = '1/' + str(int(2**i)))
    ax[1, 3].plot(w, ISn_d_r[i, :] / ISn_d_r[0, :])
    ax[2, 3].plot(w, I3n_d_r[i, :] / I3n_d_r[0, :])
    
    for j in [0, 1, 2]:
        
        for k in range(4):
    
            ax[j, k].set_xlim(180, 300)
            ax[j, k].set_ylim(0.80, 1.15)

leg = ax[0, 0].legend(framealpha = 1, loc = 1, handletextpad = 1, prop = {'size': 7.0})

ax[0, 0].set_ylabel(r'$\bar{I}_\mathrm{frac} / \bar{I}_\mathrm{full}: hydro$', fontsize = 15)
ax[1, 0].set_ylabel(r'$\bar{I}_\mathrm{frac} / \bar{I}_\mathrm{full}: ssd$', fontsize = 15)
ax[2, 0].set_ylabel(r'$\bar{I}_\mathrm{frac} / \bar{I}_\mathrm{full}: 300G$', fontsize = 15)

ax[0, 2].set_ylabel(r'$\sigma_{I_\mathrm{frac}} / \sigma_{I_\mathrm{full}}: hydro$', fontsize = 15)
ax[1, 2].set_ylabel(r'$\sigma_{I_\mathrm{frac}} / \sigma_{I_\mathrm{full}}: ssd$', fontsize = 15)
ax[2, 2].set_ylabel(r'$\sigma_{I_\mathrm{frac}} / \sigma_{I_\mathrm{full}}: 300G$', fontsize = 15)

for k in range(4):

    ax[2, k].set_xlabel('Wavelength, nm', fontsize = 15)
    ax[2, k].set_xlabel('Wavelength, nm', fontsize = 15)

plt.savefig('./stat_comp/nlte_2.pdf', bbox_inches = 'tight')

fig, ax = plt.subplots(nrows = 3, ncols = 4, figsize = (15, 10))

fig.suptitle('LTE', y = 1.005)

fig.tight_layout(w_pad = 3)

for k in [0, 2]: ax[0, k].set_title('nonrandom')
for k in [1, 3]: ax[0, k].set_title('random')

for i in range(1, Nfrac):
    
    ax[0, 0].plot(w, I0l_m_n[i, :] / I0l_m_n[0, :], label = '1/' + str(int(2**i)))
    ax[1, 0].plot(w, ISl_m_n[i, :] / ISl_m_n[0, :])
    ax[2, 0].plot(w, I3l_m_n[i, :] / I3l_m_n[0, :])
    
    ax[0, 1].plot(w, I0l_m_r[i, :] / I0l_m_r[0, :], label = '1/' + str(int(2**i)))
    ax[1, 1].plot(w, ISl_m_r[i, :] / ISl_m_r[0, :])
    ax[2, 1].plot(w, I3l_m_r[i, :] / I3l_m_r[0, :])
    
    ax[0, 2].plot(w, I0l_d_n[i, :] / I0l_d_n[0, :], label = '1/' + str(int(2**i)))
    ax[1, 2].plot(w, ISl_d_n[i, :] / ISl_d_n[0, :])
    ax[2, 2].plot(w, I3l_d_n[i, :] / I3l_d_n[0, :])
    
    ax[0, 3].plot(w, I0l_d_r[i, :] / I0l_d_r[0, :], label = '1/' + str(int(2**i)))
    ax[1, 3].plot(w, ISl_d_r[i, :] / ISl_d_r[0, :])
    ax[2, 3].plot(w, I3l_d_r[i, :] / I3l_d_r[0, :])
    
    for j in [0, 1, 2]:
        
        for k in range(4):
    
            ax[j, k].set_xlim(180, 300)
            ax[j, k].set_ylim(0.80, 1.15)

leg = ax[0, 0].legend(framealpha = 1, loc = 1, handletextpad = 1, prop = {'size': 7.0})

ax[0, 0].set_ylabel(r'$\bar{I}_\mathrm{frac} / \bar{I}_\mathrm{full}: hydro$', fontsize = 15)
ax[1, 0].set_ylabel(r'$\bar{I}_\mathrm{frac} / \bar{I}_\mathrm{full}: ssd$', fontsize = 15)
ax[2, 0].set_ylabel(r'$\bar{I}_\mathrm{frac} / \bar{I}_\mathrm{full}: 300G$', fontsize = 15)

ax[0, 2].set_ylabel(r'$\sigma_{I_\mathrm{frac}} / \sigma_{I_\mathrm{full}}: hydro$', fontsize = 15)
ax[1, 2].set_ylabel(r'$\sigma_{I_\mathrm{frac}} / \sigma_{I_\mathrm{full}}: ssd$', fontsize = 15)
ax[2, 2].set_ylabel(r'$\sigma_{I_\mathrm{frac}} / \sigma_{I_\mathrm{full}}: 300G$', fontsize = 15)

for k in range(4):

    ax[2, k].set_xlabel('Wavelength, nm', fontsize = 15)
    ax[2, k].set_xlabel('Wavelength, nm', fontsize = 15)

plt.savefig('./stat_comp/lte_2.pdf', bbox_inches = 'tight')
```
